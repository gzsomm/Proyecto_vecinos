```python
df_peliculas["title"] = df_peliculas["title"].str.strip().str.lower()
df_creditos["title"] = df_creditos["title"].str.strip().str.lower()

df_peliculas.head()
```

```python
import sqlite3

# Conexión 
con = sqlite3.connect("../data/Sistema_recomendacion_peliculas.db")

# Guardamos las dos tablas desde pandas
df_peliculas.to_sql("peliculas", con, if_exists="replace", index=False)
df_creditos.to_sql("creditos", con, if_exists="replace", index=False)
```

```python
# Creamos un cursor para ejecutar SQL
cursor = con.cursor()

# Ejecutamos una consulta JOIN para unir por la columna 'title'
query = """
SELECT 
    c.movie_id,
    p.title,
    p.overview,
    p.genres,
    p.keywords,
    c.cast,
    c.crew
FROM peliculas AS p
JOIN creditos AS c ON p.title = c.title
"""

# Ejecutamos y convertimos el resultado a DataFrame
df_combinado = pd.read_sql_query(query, con)

# Mostramos las primeras filas
print(df_combinado.head())
```

```python
# Filtrar las columnas deseadas
columnas_deseadas = ['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']
df_final = df_combinado[columnas_deseadas]

# Mostrar las primeras filas
df_final.head()
```

```python
# Transformar datos de formato JSON
import json

def load_json_safe(json_str, default_value = None):
    try:
        return json.loads(json_str)
    except (TypeError, json.JSONDecodeError):
        return default_value
    
df_final["genres"] = df_final["genres"].apply(lambda x: [item["name"] for item in json.loads(x)] if pd.notna(x) else None)
df_final["keywords"] = df_final["keywords"].apply(lambda x: [item["name"] for item in json.loads(x)] if pd.notna(x) else None)
df_final["cast"] = df_final["cast"].apply(lambda x: [item["name"] for item in json.loads(x)][:3] if pd.notna(x) else None)
df_final["crew"] = df_final["crew"].apply(lambda x: " ".join([crew_member['name'] for crew_member in load_json_safe(x) if crew_member['job'] == 'Director']))
df_final["overview"] = df_final["overview"].apply(lambda x: [x])

df_final.head()
```

```python
# Convertimos 'overview' a lista de un solo string
df_final["overview"] = df_final["overview"].apply(lambda x: [str(x)])

# Convertimos los elementos de las otras columnas a string
df_final["genres"] = df_final["genres"].apply(lambda x: [str(genre) for genre in x])
df_final["keywords"] = df_final["keywords"].apply(lambda x: [str(keyword) for keyword in x])
df_final["cast"] = df_final["cast"].apply(lambda x: [str(actor) for actor in x])
df_final["crew"] = df_final["crew"].apply(lambda x: [str(crew_member) for crew_member in x])

# Combinamos todas las listas en una sola columna
df_final["tags"] = (
    df_final["overview"] +
    df_final["genres"] +
    df_final["keywords"] +
    df_final["cast"] +
    df_final["crew"]
)

# Convertimos la lista a un solo string con espacios
df_final["tags"] = df_final["tags"].apply(lambda x: " ".join(x))

# Eliminamos las columnas que ya no necesitamos
df_final.drop(columns=["genres", "keywords", "cast", "crew", "overview"], inplace=True)

# Verificamos el resultado
print(df_final.iloc[0].tags)
```

```python
# Escalamos los datos
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizar = TfidfVectorizer()
tfidf_matrix = vectorizar.fit_transform(df_final["tags"])
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000, stop_words="english")
vectors = cv.fit_transform(df_final["tags"]).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.neighbors import NearestNeighbors

model = NearestNeighbors(n_neighbors = 6, algorithm = "brute", metric = "cosine")
model.fit(tfidf_matrix)

# Asegúrate de tener esto hecho antes
df_final["title"] = df_final["title"].str.lower()
similarity = cosine_similarity(vectors)

def recomendacion(pelicula):
    pelicula = pelicula.lower()  # <- Convierte la entrada del usuario a minúsculas
    try:
        id_pelicula = df_final[df_final["title"] == pelicula].index[0]
        distances = similarity[id_pelicula]
        lista_peliculas = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
        
        print(f"\nPelículas recomendadas:")
        for i in lista_peliculas:
            print(df_final.iloc[i[0]].title.title())  # .title() para mostrar bonito
    except IndexError:
        print(f"No se encontró la película '{pelicula}' en el conjunto de datos.")
```
